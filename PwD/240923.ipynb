```python
data_x = [i for i in range(100)]
data_y = [i*i for i in data_x]
plt.xlabel("x")
plt.ylabel("f(x) = x^2")
plt.plot(data_x, data_y)
plt.show()
```

```python
data_x = [i for i in range(100)]
data_y1 = [i*i for i in data_x]
data_y2 = [i**2.1 for i in data_x]
plt.xlabel("x")
plt.ylabel("f(x) = x^2")
plt.plot(data_x, data_y1)
plt.plot(data_x, data_y2)
plt.show()
```

```python
data_x = [i for i in range(100)]
data_y1 = [i*i for i in data_x]
data_y2 = [i**2.1 for i in data_x]
plt.xlabel("x")
plt.ylabel("y")
plt.plot(data_x, data_y1, label='f(x) = x ^ 2')
plt.plot(data_x, data_y2, label='f(x) = x ^ 2.1')
plt.legend()
plt.show()
```

```python
data_x = [i for i in range(100)]
data_y1 = [i*i for i in data_x]
data_y2 = [i**2.1 for i in data_x]
plt.xlabel("x")
plt.ylabel("y")
plt.plot(data_x, data_y1)
plt.plot(data_x, data_y2)
plt.legend(['f(x)=x^2', 'g(x)=x^2.1'])
plt.show()
```

```python
data_x = [i for i in range(100)]
data_y1 = [i*i for i in data_x]
data_y2 = [i**2.1 for i in data_x]
plt.xlabel("x")
plt.ylabel("y")
plt.bar(data_x, data_y1, label='f(x) = x ^ 2')
plt.legend()
plt.show()
```
