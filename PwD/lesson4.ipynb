### Файл с данными trades.csv
#### 1. Найти среднее, стандартное отклонение, медиану цен открытия, закрытия, low и high.
#### 2. Построить слотбец со значением close-open. Найти его среднее, стандартное отклонение, медиану
#### 3. Найти записи о торговых сессиях, в которых относительное изменение цены (отношение изменения цены к open) > 0.01 (как с +, так и с -)
#### 4. Предположим, что трейдер начинает с 1000. Сколько максимум он смог бы заработать за одну ТС?
#### 5. Предположим, что трейдер начинает с 1000. Сколько максимум он смог бы заработать за 5 подряд идущих ТС?

```python
dataframe_trades = pd.read_csv('trades.csv')

dataframe_trades.head()
```

```python
# 1. Найти среднее, стандартное отклонение, медиану цен открытия, закрытия, low и high.
statistics = {
    'mean': dataframe_trades[['open', 'close', 'low', 'high']].mean(),
    'std_dev': dataframe_trades[['open', 'close', 'low', 'high']].std(),
    'median': dataframe_trades[['open', 'close', 'low', 'high']].median()
}

print("Среднее:\n", *statistics['mean'], '\n')
print("Стандартное отклонение:\n", *statistics['std_dev'], '\n')
print("Медиана:\n", *statistics['median'], '\n', '\n')

# 2. Построить слотбец со значением close-open. Найти его среднее, стандартное отклонение, медиану
dataframe_trades['change'] = dataframe_trades['close'] - dataframe_trades['open']

print(dataframe_trades[['open', 'close', 'change']].head(), '\n')

# 3. Найти записи о торговых сессиях, в которых относительное изменение цены 
# (отношение изменения цены к open) > 0.01 (как с +, так и с -)
dataframe_trades['relative_change'] = dataframe_trades['change'] / dataframe_trades['open']
filtered_trades = dataframe_trades[(dataframe_trades['relative_change'] > 0.01) | 
                                   (dataframe_trades['relative_change'] < -0.01)]

print(filtered_trades[['open', 'change', 'relative_change']].head(), '\n', len(filtered_trades), '\n')

# 4. Предположим, что трейдер начинает с 1000. Сколько максимум он смог бы заработать за одну ТС?
max_relative_change = dataframe_trades['relative_change'].max()

print(1000 * max_relative_change, '\n')

# 5. Предположим, что трейдер начинает с 1000. Сколько максимум он смог бы заработать за 5 подряд идущих ТС?
max_relative_change_5_sessions = dataframe_trades['relative_change'].rolling(window=5).sum().max()

print(1000 * max_relative_change_5_sessions)
```
